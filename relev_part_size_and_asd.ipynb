```python
def extract_parts(json_path):
    with open(json_path, 'r') as file:
        data = json.load(file)
        for shape in data['shapes']:
            parts_set.add(shape['label'])
```

```python
def process_json(json_path, class_label):
    with open(json_path, 'r') as file:
        data = json.load(file)
        row = {part: 0 for part in list(parts_set)}  # 초기 값 0
        row['class'] = class_label
        row['name'] = json_path.split('/')[-1].split('.')[0]
        # if class_label == 'ASD':
        #     row['class'] = 0
        # else:
        #     row['class'] = 1

        image_height = data['imageHeight']
        image_width = data['imageWidth']
        image_area = image_height * image_width

        for shape in data['shapes']:
            part = shape['label']
            points = shape['points']
            x1, y1 = points[0]
            x2, y2 = points[1]
            part_area = (x2 - x1) * (y2 - y1)
            part_ratio = part_area * 100 / image_area
            row[part] = part_ratio
        df.loc[len(df)] = row
```

```python
for c in classes:
    path = os.path.join(base_path, c)
    for json_file in os.listdir(path):
        if json_file.endswith('.json'):
            extract_parts(os.path.join(path, json_file))

columns = ['name'] + ['class'] + list(parts_set)
df = pd.DataFrame(columns=columns)

for c in classes:
    path = os.path.join(base_path, c)
    for json_file in os.listdir(path):
        if json_file.endswith('.json'):
            process_json(os.path.join(path, json_file), c)
            break
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# df.to_csv('part_proportion.csv', index=False)
# df
```

## Classification 수행

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, mean_squared_error, r2_score

df['class'] = df['class'].map({'ASD': 0, 'TD': 1})

X = df.drop(columns=['class'])
y = df['class']

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

clf_svm = SVC(random_state=42)
clf_svm.fit(X_train_class, y_train_class)
y_pred_class_svm = clf_svm.predict(X_test_class)

clf_rf = RandomForestClassifier(random_state=42)
clf_rf.fit(X_train_class, y_train_class)
y_pred_class_rf = clf_rf.predict(X_test_class)

print("\nSVM Classification Report:")
print(classification_report(y_test_class, y_pred_class_svm))
print("Accuracy:", accuracy_score(y_test_class, y_pred_class_svm))
print("\nRandom Forest Classification Report:")
print(classification_report(y_test_class, y_pred_class_rf))
print("Accuracy:", accuracy_score(y_test_class, y_pred_class_rf))
```

```python
def merge_columns(df, left_col, right_col, new_col_name):
    # df[new_col_name1] = df[left_col] + df[right_col]
    df[new_col_name] = abs(df[left_col] - df[right_col])
    df.drop(columns=[left_col, right_col], inplace=True)

merge_columns(df, 'left_hand', 'right_hand', 'hand_diff')
merge_columns(df, 'left_foot', 'right_foot', 'foot_diff')
merge_columns(df, 'left_eye', 'right_eye', 'eye_diff')
merge_columns(df, 'left_ear', 'right_ear', 'ear_diff')
merge_columns(df, 'left_arm', 'right_arm', 'right_diff')
columns_to_drop = ['name','body']
df.drop(columns=columns_to_drop, inplace=True)
df
```

## classification

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import classification_report, accuracy_score, mean_squared_error, r2_score


df['class'] = df['class'].map({'ASD': 0, 'TD': 1})

X = df.drop(columns=['class'])
y = df['class']

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


clf_lr = LogisticRegression(random_state=42)
clf_lr.fit(X_train_class, y_train_class)
y_pred_class_lr = clf_lr.predict(X_test_class)

clf_rf = RandomForestClassifier(random_state=42)
clf_rf.fit(X_train_class, y_train_class)
y_pred_class_rf = clf_rf.predict(X_test_class)


print("\nLogistic Regression Classification Report:")
print(classification_report(y_test_class, y_pred_class_lr))
print("Accuracy:", accuracy_score(y_test_class, y_pred_class_lr))

print("\nRandom Forest Classification Report:")
print(classification_report(y_test_class, y_pred_class_rf))
print("Accuracy:", accuracy_score(y_test_class, y_pred_class_rf))
```
